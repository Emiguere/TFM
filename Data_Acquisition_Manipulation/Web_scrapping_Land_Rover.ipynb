```python
import bs4 #Beautiful Soup
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

## <div align="center"> Land Rover 2010-2013 </div>

##### CSV file

```python
headers='vehicle;  km; year; power; transmision; fuel_type; consume; country;emisions;price \n'
filename='land_rover_2010_2013.csv'
f=open(filename,'w')
f.write(headers)
```

##### Data acquisition

```python
#https://www.autoscout24.es/lst/land-rover?sort=age&desc=1&ustate=N%2CU&size=20&page=1&cy=E&fregto=2013&fregfrom=2010&atype=C&
my = 'https://www.autoscout24.es/lst/land-rover?sort=age&desc=1&ustate=N%2CU&size=20&page='
page = '&cy=E&fregto=2013&fregfrom=2010&atype=C&'
```

```python
#For each i, we'll read a page with 20 cars, total 400 cars.
for i in range(1,21):
    #page 
    print (i)
    page_i = my+str(i)+page
    #print (page_i)
    
    #Open the conexion
    uClient = uReq(page_i)
    #Save the html code
    page_html = uClient.read()
    #close the conexion
    uClient.close()
    #Parse
    page_soup = soup(page_html,'html.parser')
    
    #List of cars
    cars = page_soup.findAll("div",{"class":"cldt-summary-full-item"})
    
    #We'll get the features for each car
    for car in cars:
        vehicle = car.div.div.div.a.div.div.h2.text
        price = car.div.findAll('div',{'class':'cldt-summary-payment'})[0].text.strip()[0:8]
        km = car.div.findAll('div',{'class':'cldt-summary-vehicle-data'})[0].li.text.strip()
        year = car.div.findAll('div',{'class':'cldt-summary-vehicle-data'})[0].ul.findAll('li')[1].text.strip()
        power = car.div.findAll('div',{'class':'cldt-summary-vehicle-data'})[0].ul.findAll('li')[2].text.strip()
        transmision = car.div.findAll('div',{'class':'cldt-summary-vehicle-data'})[0].ul.findAll('li')[5].text.strip()
        fuel_type = car.div.findAll('div',{'class':'cldt-summary-vehicle-data'})[0].ul.findAll('li')[6].text.strip()
        consume = car.div.findAll('div',{'class':'cldt-summary-vehicle-data'})[0].ul.findAll('li')[7].text.strip()[0:4]
        ##country = car.findAll('div',{'class':'cldt-summary-seller'})[0].div.find('div',{'class':'cldf-summary-seller-contact-address'}).text
        try:
            country = car.findAll('div',{'class':'cldt-summary-seller'})[0].div.findAll('div',{'class':'cldf-summary-seller-contact-address'})[0].text.strip()[0:2]
        except:
            country = 'Unknown'
        emisions = car.div.findAll('div',{'class':'cldt-summary-vehicle-data'})[0].ul.findAll('li')[8].text.strip()[0:4]  
        price = car.div.findAll('div',{'class':'cldt-summary-payment'})[0].text.strip()[0:8]
        #Writing the csv
        f.write(vehicle +';'+ km +';'+ year +';'+ power +';'+ transmision +';'+ fuel_type +';'+ consume +';'+ country +';'+ emisions +';'+ price +'\n')
```

```python
f.close()
```

# <div align="center"> Data Manipulation

##### Imports

```python
import pandas as pd
pd.set_option('display.max_rows',3000)
```

##### Reading csv's

```python
df_10_13 = pd.read_csv('land_rover_2010_2013.csv',sep=';',encoding='latin-1')
df_14_15 = pd.read_csv('land_rover_2014_2015.csv',sep=';',encoding='latin-1')
df_16 = pd.read_csv('land_rover_2016.csv',sep=';',encoding='latin-1')
df_17 = pd.read_csv('land_rover_2017.csv',sep=';',encoding='latin-1')
df_18_19 = pd.read_csv('land_rover_2018_2019.csv',sep=';',encoding='latin-1')
df_20 = pd.read_csv('land_rover_2020.csv',sep=';',encoding='latin-1')
```

```python
df = pd.concat([df_10_13,df_14_15,df_16,df_17,df_18_19,df_20],axis = 0)
df=df.reset_index()
```

```python
df
```

##### Creating new features

```python
df['Brand']= df['vehicle'].str.split().str[0]+ ' '+df['vehicle'].str.split().str[1]
df['Model']=df['vehicle'].str.split().str[2:].str.join(sep=',').str.replace(',',' ')
df['Km'] = df['  km'].str.split().str[0].str.replace('.','')
df['Power'] = df[' power'].str.split().str[2].str[1:]
df['Price'] = df['price '].str.split().str[1].str.replace(',','').str.replace('.','')
#df['Year'] = pd.to_datetime(df[' year'])
```

#### Removing unnecessary columns

```python
df = df.drop(columns=[' consume',' country','emisions','price ',' power','  km','vehicle'])
df= df[['Brand', 'Model',' year','Km','Power',' transmision', ' fuel_type', 'Price']]
```

##### Filtering the dataframe

```python
df.shape
```

```python
#droping duplicates
df = df.drop_duplicates(keep='first')
df.shape
```

#### Models

```python
unique_values = df['Model'].unique()
print(unique_values)
```

```python
#Removing wrong models or not interesting models
df = df[df['Model']!='']
df = df[df['Model']!='Series']
df = df[df['Model']!='Freelander']
```

```python
#Land Rover models
unique_values = df['Model'].unique()
print(unique_values)
```

```python
#
df = df[df[' year']!='-/- (Año)']
df['Years']= df[' year'].str.replace('ene-20','2020')
df['Years']= df['Years'].str.replace('may-20','2020')
df['Years']= df['Years'].str.replace('jul-20','2020')
df['Years']= df['Years'].str.replace('abr-20','2020')
df['Years']= df['Years'].str.replace('dic-20','2020')
df['Years']= df['Years'].str.replace('jun-20','2020')
df['Years']= df['Years'].str.replace('feb-20','2020')
df['Years']= df['Years'].str.replace('ago-20','2020')
df['Years']= df['Years'].str.replace('mar-20','2020')
df['Years'] = pd.to_datetime(df['Years'])
df['Year'] = df['Years'].dt.year
df = df.drop(columns=[' year'])
df = df.drop(columns=['Years'])
```
